Step 1: Import libs

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler, PowerTransformer, QuantileTransformer
from matplotlib import pyplot as plt
```

Step 2: Load dataset

```python
# load dataset
df = pd.read_csv('pima-indians-diabetes.csv', header=None)
```

Step 3: Check data

```python
# view data
print(df.head())

# statistic data
print(df.describe())

# check type of data
print(df.dtypes)
```

Step 4: define a function to tranform

```python
# declare a function  to transform with para method
def i_transform(df, method):
    i_trans= method
    # fit data
    df_trans = i_trans.fit_transform(df)

    # convert transformed data to dataframe

    df_trans_i =pd.DataFrame(df_trans)
    return df_trans_i
```

Step 5: Define a function to visualize

```python
def visualize_data(data):
    fig = data.hist()
    plt.show
```

Step 6: Combine some transformation method

```python
# declare trans_method array to store transformation methods
trans_method = [MinMaxScaler(),  StandardScaler(), RobustScaler(), PowerTransformer(), QuantileTransformer()]

for i_order in trans_method:
    # apply i_transform method
    df_trans = i_transform(df, i_order)

    # visualization i_method
    print(visualize_data(df_trans))  
```
